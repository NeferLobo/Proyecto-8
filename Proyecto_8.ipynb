```python
#Parte 2

import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("casasboston.csv")
df = datos[["RM","CRIM", "MEDV", "TOWN", "CHAS"]]

df = datos.rename(columns={
	"TOWN":"CIUDAD",
	"CRIM":"INDICE_CRIMEN",
	"INDUS":"PCT_ZONA_INDUSTRIAL",
	"CHAS":"RIO_CHARLES",
	"RM":"N_HABITACIONES_MEDIO",
	"MEDV":"VALOR_MEDIANO",
	"LSTAT":"PCT_CLASE_BAJA"
})
df.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
```

```python
df.plot.scatter(x="INDICE_CRIMEN", y="VALOR_MEDIANO", alpha=0.2)
plt.show()
```

```python
df.plot.scatter(y="INDICE_CRIMEN", x="VALOR_MEDIANO", alpha=1)
plt.show()
```

```python
df.plot.scatter(y="INDICE_CRIMEN", x="VALOR_MEDIANO", alpha=0.5)
plt.show()
```

```python
df.plot.scatter(y="INDICE_CRIMEN", x="VALOR_MEDIANO", alpha=0.25)
plt.show()
```

```python
valor_por_ciudad = df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
